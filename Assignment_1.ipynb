```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
```

```python
data = pd.read_csv("C:/Users/onkar/Downloads/Assignment/Assignment/Project - 1&2_ Simple_Reg/Project - 1&2_ Simple_Reg/delivery_time.csv")
print(data.head(5))
```

```python
print("Missing data in Delivery_time -",data.isnull().sum())
```

```python
print("Delivery Data Summary:",data.describe())
```

```python
sns.scatterplot(x = 'Sorting Time', y = 'Delivery Time',data = data)
plt.title("Delivery Time vs Sorting Time")
plt.savefig('Delivery_Time_vs_Sorting_Time_prj1.png')
plt.show()
```

```python
sns.pairplot(data)
plt.savefig('data_prj1.png')
plt.show()
```

```python
X = data[['Sorting Time']]
Y = data['Delivery Time']
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size=0.10)
```

```python
model_del = LinearRegression()
```

```python
model_del.fit(X_train,Y_train)
```

```python
prd_del = model_del.predict(X_test)
print("Model Prediction complete")
```

```python
r2_delivery = r2_score(Y_test,prd_del)
r2_delivery
```

```python
Mean_sqrd_del = mean_squared_error(Y_test,prd_del)
Mean_sqrd_del
```

```python
print(f"Salary Hike Model MSE: {Mean_sqrd_del:.2f}")
print(f"Salary Hike Model R2: {r2_delivery*100:.2f}")
```
