```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
```

```python
data = pd.read_csv("C:/Users/onkar/Downloads/Assignment/Assignment/Project - 1&2_ Simple_Reg/Project - 1&2_ Simple_Reg/salary_data.csv")
print(data.head(5))
```

```python
print("Missing data in salary_data -",data.isnull().sum())
```

```python
print("Salary Data Summary:",data.describe())
```

```python
sns.scatterplot(x = 'YearsExperience', y = 'Salary',data = data)
plt.title("Experiance vs Salary Hike")
plt.savefig('Experiance_vs_Salary_Hike_PRJ2.png')
plt.show()
```

```python
sns.pairplot(data)
plt.savefig('dataPRJ2.png')
plt.show()
```

```python
x = data[['YearsExperience']]
y = data['Salary']
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2)
```

```python
model_sal = LinearRegression()
model_sal.fit(x_train,y_train)
```

```python
pred_sal = model_sal.predict(x_test)
print("Model Prediction complete")
```

```python
r2_salary = r2_score(y_test,pred_sal)
r2_salary
```

```python
mean_sqrd_sal = mean_squared_error(y_test,pred_sal)
mean_sqrd_sal
```

```python
print(f"Salary Hike Model MSE: {mean_sqrd_sal:.2f}")
print(f"Salary Hike Model R2: {r2_salary*100:.2f}")
```
